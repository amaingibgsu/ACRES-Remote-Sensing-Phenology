```python
import matplotlib.pyplot as plt

# parameter setup for quality of plot
plt.rcParams.update({'font.size': 12, 'font.family': 'sans-serif'})
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24, 8))

# PANEL A: Overall Suitability (Donut Chart)
labels = ['Fully Suitable\n(n=70)', 'Partially Suitable\n(n=55)']
sizes = [56, 44]
colors = ['#2ca02c', '#ff7f0e'] # Green and Orange
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, pctdistance=0.85)
ax1.add_artist(plt.Circle((0,0), 0.70, fc='white')) # Create the 'Donut' hole
ax1.set_title("A. Overall RS Operational Readiness", fontsize=16, fontweight='bold', pad=20)

# PANEL B: Phenology Peak (Line Chart)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
flowering_counts = [0, 2, 8, 25, 55, 95, 92, 60, 30, 10, 2, 0] # Based on Section 3.11
ax2.plot(months, flowering_counts, color='#e74c3c', marker='o', linewidth=3, label='Flowering Species')
ax2.axvspan('Jun', 'Jul', color='yellow', alpha=0.3, label='Optimal RS Window')
ax2.set_ylabel('Number of Species', fontsize=14)
ax2.set_title("B. Optimal RS Sensing Window", fontsize=16, fontweight='bold', pad=20)
ax2.legend()
ax2.grid(True, linestyle='--', alpha=0.5)

# PANEL C: Scorecard Hierarchy (Horizontal Bar Chart)
scorecards = ['Rough Grazing', 'Winterage', 'Grassland', 'Peatland', 'Woodland', 
              'Coastal Grassland', 'Grassland on Peat Soil', 'Corncrake', 'Breeding Wader', 'Chough']
suitability = [23, 27, 33, 45, 62.5, 64, 67, 69, 77, 91]
bar_colors = plt.cm.RdYlGn([val/100 for val in suitability])
bars = ax3.barh(scorecards, suitability, color=bar_colors, edgecolor='black')

# Add precise labels to each of the bars
for bar in bars:
    ax3.text(bar.get_width() + 1, bar.get_y() + bar.get_height()/2, f'{bar.get_width()}%', va='center', fontweight='bold')

ax3.set_xlim(0, 110)
ax3.set_xlabel('Operational Suitability (%)', fontsize=14)
ax3.set_title("C. Readiness by Habitat Scorecard", fontsize=16, fontweight='bold', pad=20)

plt.tight_layout()
plt.savefig(r"C:\Users\Alex.Maingi\Downloads\Visual_Abstract_ACRES_Final.png", dpi=600, bbox_inches='tight')
#plt.savefig("Visual_Abstract_ACRES_Final.png", dpi=600)
plt.show()
```
